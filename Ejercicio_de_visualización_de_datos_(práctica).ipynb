```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('/The_World_s_Highest_Paid_Athletes.xlsx')
```

Los deportistas más pagados por deporte

```python
df.head()
```

```python
Sports = df.groupby('Sport')['Total Earnings'].sum()
plt.bar(Sports.index, Sports.values)
plt.xticks(rotation = 45);
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.ylabel('Millones de dólares', fontsize = 12)
plt.xlabel('Deportes', fontsize = 12)
plt.title('Ganancias totales de los 50 deportistas más pagados del mundo por deporte')
plt.grid(axis = 'y', alpha = 0.7, linestyle= '-')
```
